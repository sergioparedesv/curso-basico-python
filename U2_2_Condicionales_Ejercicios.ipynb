# Ejercicio de Condicionales

**1.** Crea un programa que pueda indicar si un **número** es positivo o negativo.

**2.** Diego recibe **10.000 pesos** y es enviado al supermercado a comprar algunos productos de aseo, la cuenta puede salir de **8.000 a 12.000 pesos**, genera un programa que le diga si le alcanza, y, si lo hace, indique cuál es el **cambio** que debe recibir Diego.

**3.** Crea un programa que calcule el **descuento** que se le puede hacer a un cliente a partir de su compra, si el cliente compra 1000 lápices, podrá hacérsele un descuento del **15%**, si pide menos de 1000 lápices solo podrá pedirle un descuento del **10%**.

**4.** Crearemos un programa que reciba **tres parámetros** y nos diga cuál es el mayor, utilizando la **condición** `if y else`

**5.** Crea un programa que contenga diferentes rutinas de ejercicio a la semana; dependiendo del día que el usuario escoja, el programa le recomendará la rutina:

 - **Lunes**: Flexiones 
 - **Martes**: Sentadillas
 - **Miércoles**: Abdominales
 - **Jueves**: Saltar lazo
 - **Viernes** : Natación

¡Para otros días el programa recomendará al usuario descansar!

**6.** Crea un programa que le ayude al usuario a calcular los problemas fisiológicos que puede tener al estar en diferentes alturas siguiendo la siguiente tabla:


| Altura             | Efectos |
|--------------------|---------|
| 0 a 1000 metros    | No se presentan efectos fisiológicos en reposo ni en ejercicio físico.       |
| 1001 a 2000 metros | Ligeros efectos durante la actividad deportiva, disminución del ejercicio físico.      |
| 2000 a 5000 metros | Se puede observar modificaciones fisiológicas en reposo y muy acentuadas durante el ejercicio.     |
| 5000 o más metros  | El efecto negativo es muy alto y la vida permanente es imposible.    |

**7.** Retomando al ejercicio de los **semáforos**, incluiremos una variante que verifique que si ninguna de las 3 luces está prendida, le imprima al usuario que el funcionamiento del semáforo no se encuentra disponible.


**8.** Retomando el programa de la **edad**, esta vez verificarás que la edad sea positiva y luego que verifique si es mayor de **edad.**

**9.** Siguiendo con los ejercicios sobre la **edad**, crea un programa que, al insertar el año de nacimiento y luego comprobando si ya cumplió años o no, **calcule su edad.**

**10.** Tenemos una caja fuerte con una **contraseña *6421***, crea un programa con una **sentencia** **`and`** en la que pida que se inserten los cuatro números que son la contraseña de la caja fuerte.

**11.** Crea un programa que, utilizando la sentencia `elif`, compruebe si un número ingresado es **divisible por 2 y por 3**.

**12.** Crea un programa que, utilizando la **sentencia `or`**, indique si alguien tiene **cédula de ciudadanía, cédula de extranjería o pasaporte**, para poder ingresar a un bar.
